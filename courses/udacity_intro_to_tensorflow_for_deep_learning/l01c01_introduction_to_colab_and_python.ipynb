```
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import torch
import torchaudio

# Charger le modèle et le processeur
# processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-xlsr-53")
# model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-xlsr-53")
# Load model directly
from transformers import AutoProcessor, AutoModelForPreTraining

processor = AutoProcessor.from_pretrained("facebook/wav2vec2-xls-r-300m")
model = AutoModelForPreTraining.from_pretrained("facebook/wav2vec2-xls-r-300m")


# Charger un fichier audio
speech_array, sampling_rate = torchaudio.load("votre_audio.wav")
speech = torchaudio.transforms.Resample(orig_freq=sampling_rate, new_freq=16000)(speech_array).squeeze()

# Préparer les données pour le modèle
input_values = processor(speech, sampling_rate=16000, return_tensors="pt", padding=True).input_values

# Faire une prédiction
with torch.no_grad():
    logits = model(input_values).logits

# Décoder la prédiction
predicted_ids = torch.argmax(logits, dim=-1)
transcription = processor.batch_decode(predicted_ids)
print("Transcription :", transcription[0])
```

```
from transformers import AutoProcessor, AutoModelForCTC
import torch
import torchaudio

# Load the fine-tuned model and processor
processor = AutoProcessor.from_pretrained("facebook/wav2vec2-base-900h")
model = AutoModelForCTC.from_pretrained("facebook/wav2vec2-base-900h")

# Charger un fichier audio
speech_array, sampling_rate = torchaudio.load("votre_audio.wav")
speech = torchaudio.transforms.Resample(orig_freq=sampling_rate, new_freq=16000)(speech_array).squeeze()

# Préparer les données pour le modèle
input_values = processor(speech, sampling_rate=16000, return_tensors="pt", padding=True).input_values

# Faire une prédiction
with torch.no_grad():
    logits = model(input_values).logits

# Décoder la prédiction
predicted_ids = torch.argmax(logits, dim=-1)
transcription = processor.batch_decode(predicted_ids)
print("Transcription :", transcription[0])
```
